```python
import numpy as np
from numpy.random import default_rng
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
x = np.arange(0,20,0.1);
sin = np.sin(x);
cos = np.cos(x);
```

```python
plt.plot(x,sin);
plt.plot(x,cos);

plt.title("sin(x) vs cos(x)");
plt.xlabel("x-value");
plt.ylabel("y-value");
```

### Grid

```python
fig = plt.figure(figsize=(10,8));

ax1 = plt.subplot2grid((2,2),(0,0));
ax2 = plt.subplot2grid((2,2),(0,1));
ax3 = plt.subplot2grid((2,2),(1,0),colspan=2);

x1 = np.arange(0,10,0.5);
y1 = 3*x1+1;
ax1.plot(x1,y1,color="green");

x2 = np.arange(0,10,0.5);
y2 = 3*(x2**2) + 2*x2 + 10;
ax2.plot(x2,y2,color="orange");

x3 = np.arange(-10,10,0.5);
y3 = 3*(x3**3) + 5*(x3**2) + 10*x3 -10;
ax3.plot(x3,y3,color="blue");

rng = default_rng();
y4 = y3+rng.integers(low=-200,high=200,endpoint=True,size=40);
ax3.scatter(x3,y4,color="brown");
```

### Playing with Big O notation
- n
- nlogn
- n^2 (square)
- n^3 (cube)

```python
x = np.arange(1,1001);
n = x;
nlogn = np.empty(1000);
nsq = np.empty(1000);
ncb = np.empty(1000);
for i in x:
    nlogn[i-1] = i*np.log(i);
    nsq[i-1] = i**2;
    ncb[i-1] = i**3;
    
fig = plt.figure(figsize=(10,8));

ax1 = plt.subplot2grid((2,2),(0,0));
ax2 = plt.subplot2grid((2,2),(0,1));
ax3 = plt.subplot2grid((2,2),(1,0));
ax4 = plt.subplot2grid((2,2),(1,1));

ax1.plot(x,n,color="green");
ax2.plot(x,nlogn,color="brown");
ax3.plot(x,nsq,color="yellow");
ax4.plot(x,ncb,color="blue");
```
